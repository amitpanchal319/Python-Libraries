```python
import numpy as np
import pandas as pd
np.random.seed(1) 
no_sample = 1000
class_0_ratio = 0.9
no_class_0 = 900
no_class_1 = 100
```

```python
class_0 = {'feature1': np.random.normal(0,1, no_class_0)}
```

```python
class_0 = {'feature1': np.random.normal(0,1, no_class_0), 
           'feature2': np.random.normal(0,1, no_class_0),
           'target': [0]*no_class_0}
```

```python
class_0 = pd.DataFrame(class_0)
```

```python
class_0
```

```python
class_1 = pd.DataFrame({'feature1': np.random.normal(3,1, no_class_1), 
           'feature2': np.random.normal(3,1, no_class_1),
           'target': [1]*no_class_1})
```

```python
class_1
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df
```

```python
df.target.value_counts()
```

```python
# Upsampling 

df_minority = df[df.target == 1]
df_minority = df[df.target == 0]
```

```python
df_minority
```

```python
from sklearn.utils import resample
```

```python
resample(df_minority, replace = True, n_samples = len(df_minority), random_state = 1)
```
