```python
#Multivariate Analysis

cols = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar',
       'sodium',
       'potassium', 'haemoglobin', 'packed_cell_volume',
        'class']
g = sns.PairGrid(df[cols], palette='coolwarm')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.map_diag(sns.histplot)
g.add_legend()
g.fig.suptitle('PairGrid for selected columns', y=1.02)  # Adjust the title position
plt.show()
```

# Plotly Analysis 

```python
fig = px.scatter(df, 
                 x='age', 
                 y='blood_pressure', 
                 color='class', 
                 hover_data=['serum_creatinine','haemoglobin'], 
                 title='Interactive Scatter Plot with Hover Information')

fig.show()
```

```python
fig = px.scatter_3d(df, 
                    x='age', 
                    y='blood_pressure', 
                    z='serum_creatinine', 
                    color='class', 
                    hover_data=['serum_creatinine', 'haemoglobin'], 
                    title='3D Scatter Plot')

fig.show()
```

```python
import plotly.graph_objects as go
 
data_corr = df.corr(numeric_only=True)
fig = go.Figure(data = go.Heatmap(z=data_corr.values,
                           x = data_corr.columns, 
                           y = data_corr.index))
fig.show()
```
