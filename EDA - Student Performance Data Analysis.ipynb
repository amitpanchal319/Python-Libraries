```python
students_grades_df = df.copy()
```

```python
students_grades_df.drop(['G1', 'G2', 'G3'], axis = 1)
```

```python
def detect_outliers(columns):
    outlier_indices = []
    for column in columns:
        Q1=students_grades_df[column].quantile(.25)
        Q3=students_grades_df[column].quantile(.75)
        IQR = Q3-Q1
        # Mask is boolean condition
        mask = (students_grades_df[column] >= Q1-1.5*IQR) & (students_grades_df[column] <= Q1+1.5*IQR)
        mask = mask.to_numpy() # Bool value to numpy array
        
        # find the indices of values that do not satisfy the mask condition >> that will be outliers 
        
        false_indices = np.argwhere(~mask)
        outlier_indices.append(false_indices)
    return np.unique(np.concatenate(outlier_indices).ravel())
        
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Sample DataFrame (Assuming it includes both numerical and categorical data)
data = {
    'age': [15, 16, 16, 15, 15, 17, 18, 19, 16, 15, 14, 22],  # Including some outliers
    'absences': [3, 4, 2, 6, 8, 1, 0, 15, 12, 3, 4, 10],       # Including some outliers
    'sex': ['F', 'M', 'F', 'F', 'M', 'M', 'F', 'M', 'M', 'F', 'F', 'M'],
    'address': ['U', 'U', 'U', 'R', 'U', 'U', 'U', 'R', 'R', 'U', 'R', 'R']
}

students_grades_df = pd.DataFrame(data)

# Separate the columns
num_cols = ['age', 'absences']
cat_cols = ['sex', 'address']

# Define the transformers
numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')  # Drop first to avoid multicollinearity

# Combine the transformers into a preprocessor using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("StandardScaler", numeric_transformer, num_cols),
        ("OneHotEncoder", oh_transformer, cat_cols)
    ]
)

# Apply the preprocessor to the DataFrame
preprocessed_data = preprocessor.fit_transform(students_grades_df)

# Convert the resulting array back to a DataFrame for easier interpretation
# Note: OneHotEncoder generates multiple columns for each categorical variable, so we'll need to handle that
encoded_cat_cols = preprocessor.named_transformers_["OneHotEncoder"].get_feature_names_out(cat_cols)
all_columns = num_cols + list(encoded_cat_cols)

preprocessed_df = pd.DataFrame(preprocessed_data, columns=all_columns)

print(preprocessed_df)
```
