```python
for i in range (10):
    for j in range (5):
        print ("hello world")
```

```python
for i in range (100):
    for j in range (5):
        print ("Amit Panchal is a Great human BeING")
```

```python
for i in range(4):
    for j in range(1,13):
        if j%4 == 0:
            break
        print(j)
```

```python
for i in range(4):
    for j in range(1,13):
        if j%2 != 0:
            continue
        print(j)
```
